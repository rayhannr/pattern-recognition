## Menghitung Image  Similarity Menggunakan Cosine Similarity dan Euclidean Distance

#### Import library yang diperlukan

```python
import numpy as np
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
```

#### Inisiasi array untuk menampung gambar

```python
images = []
preprocessed_images = []
```

#### Preprocessing gambar
Preprocessing di sini melibatkan lima tahap yaitu :
1. Rotasi
2. Mengubah citra RGB ke grayscale
3. Resize gambar menjadi 303.75 x 379.6875
4. Crop gambar tepat di tengah-tengah sebesar setengah dari hasil resize
5. Filter gambar menggunakan Gaussian blur

```python
for i in range(0,5):
    image = Image.open('image{}.jpg'.format(i+1))
    
    #Storing original image to images array
    images.append(image)
    
    #Image alignment
    image = image.rotate(-45)
    
    #Convert to grayscale
    gs = image.convert(mode = 'L')
    
    width = 303.75
    height = 379.6875
    
    #Resizing
    gs.thumbnail((width, height))
    
    #Cropping
    gs = gs.crop((width/4, height/4, 3*width/4, 3*height/4))
    
    #Filtering using Gaussian blur
    gs = gs.filter(ImageFilter.GaussianBlur(radius=0.75))
    
    #Saving the preprocessed file and storing it in preprocessed_images array
    gs.save('image_gs{}.jpg'.format(i+1))
    preprocessed_images.append(gs)
```

```python
print(images[1].size, preprocessed_images[1].size)
```

### Fungsi untuk menampilkan gambar supaya memudahkan dalam membandingkan gambar asli dan hasil preprocessing

```python
def show_image(img):
    fig = plt.figure(figsize=(18,18))
    imgs = ("Image 1", img[0]), ("Image 2", img[1]), ("Image 3", img[2]), ("Image 4", img[3]), ("Image 5",img[4])

    for (i, (name, image)) in enumerate(imgs):
        axis = fig.add_subplot(1, 5, i + 1)
        axis.set_title(name)
        plt.imshow(image, cmap = plt.cm.gray)
        plt.axis("off")

    plt.show()

show_image(images)
show_image(preprocessed_images)
```

```python
image_pixel = []
for i in range(0,5):
    pixel = np.asarray(preprocessed_images[i])
    image_pixel.append(pixel)
    
print(image_pixel[3])
print(image_pixel[3].size)
```

### Fungsi untuk menghitung cosine similarity

```python
def multiply_column_sum(img1, img2):
    return np.sum(img1.astype('float') * img2.astype('float'))

def quadratic_sum(img):
    return np.sqrt(np.sum(np.square(img.astype('float'))))

def cosine_similarity(img1, img2):
    return multiply_column_sum(img1, img2) / (quadratic_sum(img1) * quadratic_sum(img2))

cosine_similarity(image_pixel[1], image_pixel[4])
```

### Fungsi untuk menghitung euclidean distance

```python
def euclidean_distance(img1, img2):
    return np.sqrt(np.sum(np.square(img1.astype("float") - img2.astype("float"))))

euclidean_distance(image_pixel[2], image_pixel[3])
```

### Fungsi untuk menampilkan dua gambar beserta tingkat kemiripannya
Semakin besar tingkat kemiripan dua buah gambar, nilai cosine similarity-nya akan semakin besar pula sedangkan nilai euclidean distance-nya semakin kecil. Jika dua gambar tersebut identik, cosine similarity = 1 dan euclidean distance = 0

```python
def compare_image(img1, img2, img1_px, img2_px):
    cossim = cosine_similarity(img1_px, img2_px)
    ecl = euclidean_distance(img1_px, img2_px)
    fig = plt.figure(figsize=(4,4))
    plt.suptitle("Cosine: %.2f, Euclidean: %.2f" % (cossim, ecl))    
    
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(img1, cmap = plt.cm.gray)
    plt.axis("off")

    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(img2, cmap = plt.cm.gray)
    plt.axis("off")
    
    plt.show()
    
compare_image(preprocessed_images[1], preprocessed_images[1], image_pixel[1], image_pixel[1])
```

### Membandingkan tingkat kemiripan masing-masing gambar

```python
for i in range(0,5):
    for j in range(i+1, 5):
        compare_image(preprocessed_images[i], preprocessed_images[j], image_pixel[i], image_pixel[j])
```

```python
img_similarity = np.zeros((5,5))
for i in range(5):
    img_similarity[i, i] = 1
    for j in range(i+1, 5):
        img_similarity[i, j] = cosine_similarity(image_pixel[i], image_pixel[j])
        img_similarity[j, i] = img_similarity[i, j]
        
img_similarity
```

```python
img_similarity = np.zeros((5,5))
for i in range(5):
    img_similarity[i, i] = 1
    for j in range(i+1, 5):
        img_similarity[i, j] = euclidean_distance(image_pixel[i], image_pixel[j])
        img_similarity[j, i] = img_similarity[i, j]
        
img_similarity
```
