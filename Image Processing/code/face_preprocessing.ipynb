```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# Load in color image for face detection
image = cv2.imread('face2.jpg')
# Convert the image to RGB colorspace
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with the bounding boxes
fig = plt.figure(figsize = (8,8))
axl = fig.add_subplot(111)
axl.set_xticks([])
axl.set_yticks([])
axl.imshow(image_rgb)
```

```python
# Convert the image to gray 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Detect faces in the image using pre-trained face dectector
face_cascade = cv2.CascadeClassifier('C:/Python/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('C:/Python/Lib/site-packages/cv2/data/haarcascade_eye.xml')
faces = face_cascade.detectMultiScale(gray_image, 1.25, 5)

# Print number of faces found
print('Number of faces detected:', len(faces))

# Get the bounding box for each detected face
for f in faces:
    x, y, w, h = [ v for v in f ]
    cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,0), 3)
    # Define the region of interest in the image  
    roi_gray = gray_image[y:y+h, x:x+w]
    roi_color = image[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

# Display the face crops
fig = plt.figure(figsize = (6,6))
axl = fig.add_subplot(111)
axl.set_xticks([])
axl.set_yticks([])
axl.imshow(image)
```

```python
# Load in color image for face detection
image = cv2.imread('face2.jpg')
# Convert the image to gray 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image using pre-trained face dectector
face_cascade = cv2.CascadeClassifier('C:/Python/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('C:/Python/Lib/site-packages/cv2/data/haarcascade_eye.xml')
faces = face_cascade.detectMultiScale(gray_image, 1.25, 5)

# Print number of faces found
print('Number of faces detected:', len(faces))

# Get the bounding box for each detected face
for f in faces:
    x, y, w, h = [ v for v in f ]
    cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,0), 3)
    # Define the region of interest in the image  
    face_crop = gray_image[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(face_crop)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(face_crop,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

# Display the face crops
cv2.imshow('Gray image', face_crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
